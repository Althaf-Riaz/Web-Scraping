```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.flipkart.com/search?q=redmi&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
page=requests.get(url)
```

```python
print(page)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup)
```

```python
phones=soup.find('div',class_="_1YokD2 _3Mn1Gg")
```

```python
print(phones)
```

```python
items=soup.find_all('div',class_="_2kHMtA")
```

```python
print(items[0].find(class_='_4rR01T').get_text())
```

```python
name=[item.find(class_='_4rR01T').get_text() for item in items]
```

```python
spec=[item.find(class_='_1xgFaf').get_text() for item in items]
```

```python
price=[item.find(class_='_30jeq3 _1_WHN1').get_text() for item in items]
```

```python
print(name)
```

```python
print(spec)
```

```python
print(price)
```

```python
redmi=pd.DataFrame(
    {
        "Model Name" : name,
        "Short Description" : spec,
        "Price" : price
    }
)
```

```python
print(redmi)
```

```python
redmi.to_csv('redmi.csv')
```
